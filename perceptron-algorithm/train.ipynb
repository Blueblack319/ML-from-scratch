```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
from model import Perceptron
```

```python
X, y = datasets.make_blobs(n_samples=200, n_features=2, centers=2, random_state=1234)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
print(X_test.shape)
print(X_test[:3])
print(y_test.shape)
```

```python
def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)
```

```python
model = Perceptron()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy(y_test, y_pred))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_test[:, 0], X_test[:, 1], marker="o", c=y_test)

# class_0이 X축
x0_1 = np.amin(X_test[:, 0])
x0_2 = np.amax(X_test[:, 0])

print(model.weight)

x1_1 = (-model.weight[0] * x0_1 - model.bias) / model.weight[1]
x1_2 = (-model.weight[0] * x0_2 - model.bias) / model.weight[1]

ax.plot([x0_1, x0_2], [x1_1, x1_2], "k")

ymin = np.amin(X_test[:, 1])
ymax = np.amax(X_test[:, 1])
ax.set_ylim([ymin - 3, ymax + 3])

plt.show()
```
