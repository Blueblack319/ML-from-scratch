```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=500, n_features=1, noise=20, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234
)

fig = plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], y, color="b", marker="o", s=30)
plt.show()
```

```python
def MSE(y_test, y_pred):
    return np.mean((y_test - y_pred)**2)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(MSE(y_test, y_pred))
```

```python
y_pred = model.predict(X)

fig = plt.figure(figsize=(8, 6))
m1 = plt.scatter(X_train, y_train, color='b', marker='o', s = 10)
m2 = plt.scatter(X_test, y_test, color='r', marker='o', s = 10)
plt.plot(X, y_pred, color='black', linewidth=2, label='Prediction')
plt.show()
```
